```python
# Bar is used to representation of frequency of discrete data variable
# Bar is apllicable on discrete data variable


import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv("D:/Data Science By SRK/RestaurantData.csv")
print(df)

plt.bar(df["sex"],df["total_bill"])
plt.xlabel("Sex")
plt.ylabel("Total Bill")
plt.title("Bar Plot on Gender & Total Bill")
plt.show
```

```python
sns.catplot(x="sex",y="total_bill",data=df)
plt.show()

sns.catplot(x="sex",y="total_bill",data=df,hue="day")
plt.show()
```

```python
sns.catplot(x="sex",y="total_bill",data=df,kind="bar")
plt.show()

# Here in graph black line on the bars indicates that the bar is formed from collecting many points(plots)
# If no black line on bar then that bar is formed from only one point(plot)
```

```python
sns.catplot(x="day",y="total_bill",data=df,kind="bar",hue="size")
plt.show()
```

```python
sns.catplot(x="smoker",y="tip",data=df,kind="bar",hue="day")
plt.show()
```
