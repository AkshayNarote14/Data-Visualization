```python
# LinePlot represents the sequencial data in line formate by joining the points in sequential order
# LinePlot is useful only when we have sequencial data(time series data) (on x-axis)
# LinePlot is a bivariate so, It applicable on two continous variables when one should be in sequential order(time series form)

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv("D:\\Data Science by SRK\\RestaurantData.csv")
print(df)

plt.plot(df["total_bill"],df["tip"],marker="o",color="green") # Here, there is none time series variable so, it is not much useful here
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.title("Line Plot Graph")
plt.show()
```

```python
sns.relplot(x="total_bill",y="tip",data=df,kind="line",marker="o")
plt.show()
```
