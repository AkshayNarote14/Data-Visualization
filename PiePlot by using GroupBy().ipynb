```python
# PiePlot is used to represent the percentages of discrete variables in piechart formate
# PiePlot is univariate so, It is applicable on only one discrete variable.
# groupby() function used to make the groups of unique variables(values)
# groupby() is not useful for continous variables, it is used for discrete variables


import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv("D:\\Data Science by SRK\\RestaurantData.csv")
print(df)

df.groupby("sex").size().plot(kind="pie",autopct="%f%%")
plt.show()
```

```python
df1 = df.groupby("day").size()   # groupby() function used to make the groups of unique variables(values) 
print(df1)

df1.plot(kind="pie",autopct="%f%%")  # we can pass different plots type names to draw their perticular type of graph
plt.show()
```

```python
df1.plot(kind="bar")
plt.show()
```

```python
df1.plot(kind="line")
plt.show()
```

```python
df1.plot(kind="box")
plt.show()
```
