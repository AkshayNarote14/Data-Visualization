```python
# ScatterPlot is (used to) keeping(ploting) points on the graph
# ScatterPlot is bivariate so, it is applicable on two continous variables(columns)
# ScatterPlot is also multivariate when we have to apply on more than two variables so, 
   # it is applicable on two continous variables(columns) & one or more than one discrete variables.
# ScatterPlot is used to observe linearity, direction & strength of data distribution


import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv("D:/Data Science by SRK/RestaurantData.csv")
print(df)

plt.scatter(x=df["total_bill"],y=df["tip"],s=20,c="green")   # s = size of points(plots) in graph, c = color for points
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.title("Scanner Plot on TotalBill & Tip")
plt.show()
```

```python
sns.scatterplot(x=df["total_bill"],y=df["tip"],c="green")
```

```python
# ScatterPlot is also appkicable on more than two variables provided two variables are continous and other can be discrete variable(s)

sns.scatterplot(x=df["total_bill"],y=df["tip"],data=df,hue="sex")  # here, without data(data=df) attribute the hue attribute is not going to work
plt.show()
```

```python
male = df[df["sex"]=="Male"]    # It will take only male records seperately
female = df[df["sex"]=="Female"]  # It will take only female records seperately

# Applying scatter plot seperately to understand backend code of hue, style, size paramters of scatterplot 

plt.scatter(male["tip"],male["total_bill"])
plt.scatter(female["tip"],female["total_bill"])
```

```python
male = df[df["sex"]=="Male"]    # It will take only male records seperately
female = df[df["sex"]=="Female"]  # It will take only female records seperately

# Applying scatter plot seperately to understand backend code of hue, style, size paramters of scatterplot 

plt.scatter(male["tip"],male["total_bill"],marker="*")
plt.scatter(female["tip"],female["total_bill"],marker="+")
```
