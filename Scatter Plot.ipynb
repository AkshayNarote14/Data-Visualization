```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("tips.csv")
df
```

```python
plt.scatter(df["total_bill"],df["tip"])
plt.xlabel("Total Bill")
plt.ylabel("Tips")
plt.show()
```

```python
sns.scatterplot(x=df["total_bill"],y=df["tip"])
```

```python
sns.scatterplot(x=df["total_bill"],y=df["tip"],hue=df["sex"])
plt.title("Scatterplot on Total Bill & Tips")
plt.show()
```

```python
male = df[df["sex"]=="Male"]    # It will take only male records seperately
female = df[df["sex"]=="Female"]  # It will take only female records seperately

# Applying scatter plot seperately to understand backend code of hue, style, size paramters of scatterplot 

plt.scatter(male["tip"],male["total_bill"])
plt.scatter(female["tip"],female["total_bill"])
```

```python
male = df[df["sex"]=="Male"]    # It will take only male records seperately
female = df[df["sex"]=="Female"]  # It will take only female records seperately

# Applying scatter plot seperately to understand backend code of hue, style, size paramters of scatterplot 

plt.scatter(male["tip"],male["total_bill"],marker="*")
plt.scatter(female["tip"],female["total_bill"],marker="+")
```

```python
sns.scatterplot(x=df["tip"],y=df["total_bill"],style=df["day"])
plt.show()
```

```python
sns.scatterplot(x=df["tip"],y=df["total_bill"],size=df["day"])
```

```python
sns.scatterplot(x=df["total_bill"],y=df["tip"],data=df,hue="size",style="time",size="day") # Here, more than two variables are used two are continous variables & three are discrete variables
plt.show()
```

#### In seaborn we have scatterplot & relplot(relativeplot) to represent the distribution of points on two contionous variables.
- Scatterplot and relplot are applied on two continous varibales.
- scatterplot and relplot also used for more than two variables but third or next variables should be descrete variables, as below and above
- differnce between scatterplot and relplot is scatter plot show legend symbols inside the graph whereas relplot shows legend symbols outside the graph

```python
sns.relplot(x=df["tip"],y=df["total_bill"],hue="day",data=df)
```

```python
sns.relplot(x=df["total_bill"],y=df["tip"],data=df,hue="time",style="day",size="size") # Here, more than two variables are used two are continous variables & three are discrete variables
plt.show()
```

```python
sns.relplot(x=df["total_bill"],y=df["tip"],col=df["sex"])
plt.show()
```

```python
sns.relplot(x=df["total_bill"],y=df["tip"],col=df["day"])
plt.show()
```

```python
sns.relplot(x=df["total_bill"],y=df["tip"],col=df["size"],col_wrap=3,hue=df["sex"])    # col_wrap means it will take how many columns you want
plt.show()
```
