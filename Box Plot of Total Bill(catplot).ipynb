```python
# BoxPlot is used to check outliers are present in data or not.
# BoxPlot is univariate so, it is apply on only one continous variable(column)

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv("D:\\Data Science by SRK\\RestaurantData.csv")
print(df)

plt.boxplot(df["total_bill"])
plt.title("BoxPlot of Total Bill")
plt.ylabel("Total Bill")
plt.show()
```

```python
sns.boxplot(y = df["total_bill"])
plt.show()
print("="*100)

sns.boxplot(x= df["total_bill"])
plt.show()
```

```python
sns.boxplot(x="day",y="total_bill",data=df)
plt.show()
```

```python
sns.catplot(x="day",y="total_bill",data=df,hue="smoker",kind="box")
```
