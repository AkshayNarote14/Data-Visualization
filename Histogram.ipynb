```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("tips.csv")
df
```

```python
plt.hist(df["tip"])
plt.show()
```

```python
plt.hist(df["total_bill"],bins=6,color="green",range=(0,60))
plt.xlabel("Total_Bill")
plt.ylabel("Frequency/Count")
plt.title("Histogram on Total Bill")
plt.xticks(list(range(0,61,10)))
plt.show()
```

```python
sns.histplot(df["total_bill"],bins=6)
plt.xticks(list(range(0,61,10)))
plt.show()
```

```python
sns.histplot(df["total_bill"],kde=True)
```

```python
sns.histplot(df["tip"],bins=10)
```
