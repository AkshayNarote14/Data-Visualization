```python
# Histogram is (used to) representation of distribution of data in the graphical formate in the form of bins(bars)
# Histogram is univariate so, it is apply on a continous variable(column)

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv("D:\\Data Science by SRK\\RestaurantData.csv")
print(df)

n,bins,patches = plt.hist(df["total_bill"],bins=6,range=(0,60))
plt.bar_label(patches)
plt.xlabel("Total Bill")
plt.ylabel("Count/Frequency")
plt.title("Histogram of Total Bill")
plt.xticks(list(range(0,61,10)))
plt.show()
```

```python
sns.histplot(df["total_bill"],bins=6,kde=True)
plt.show()
```

```python
sns.displot(df["total_bill"],bins=10,kde=True)
plt.show()
```
